# Sistema de Telemetria para Decolagem Espacial

Este notebook apresenta um sistema simples de decisão para uma nave espacial, baseado em telemetria, análise de energia e verificação de segurança.

## 1. Importação de bibliotecas

A biblioteca `random` será usada para simular a leitura dos sensores da nave.

```python
import random
```

## 2. Geração dos dados de telemetria

Nesta etapa, os parâmetros da nave são simulados com valores aleatórios para representar leituras de sensores.

```python
def gerar_dados():
    return {
        "temp_interna": round(random.uniform(0, 40), 1),
        "temp_externa": round(random.uniform(-10, 50), 1),
        "integridade": random.choice([0, 1]),
        "energia": round(random.uniform(0, 100), 1),
        "pressao_principal": round(random.uniform(0, 120), 1),
        "pressao_auxiliar": round(random.uniform(0, 120), 1),
        "modulos_criticos": random.choice([0, 1])
    }
```

## 3. Verificação de lançamento

A função abaixo compara os valores obtidos com faixas seguras previamente definidas. Se qualquer condição falhar, a decolagem deve ser abortada.

```python
def avaliar_lancamento(dados):
    motivos = []

    if not (15 <= dados["temp_interna"] <= 25):
        motivos.append("Temperatura interna fora da faixa")

    if not (4 <= dados["temp_externa"] <= 35):
        motivos.append("Temperatura externa fora da faixa")

    if dados["integridade"] == 0:
        motivos.append("Falha estrutural")

    if dados["energia"] < 60:
        motivos.append("Energia insuficiente")

    if not (30 <= dados["pressao_principal"] <= 100):
        motivos.append("Pressão do tanque principal fora da faixa")

    if not (30 <= dados["pressao_auxiliar"] <= 100):
        motivos.append("Pressão do tanque auxiliar fora da faixa")

    if dados["modulos_criticos"] == 0:
        motivos.append("Falha nos módulos críticos")

    if len(motivos) == 0:
        return "PRONTO PARA DECOLAR", ["Todos os parâmetros OK"]
    else:
        return "DECOLAGEM ABORTADA", motivos
```

## 4. Análise energética

Esta etapa calcula a autonomia restante da nave com base na capacidade total, carga atual, consumo estimado na decolagem e perdas energéticas.

```python
def calcular_autonomia(energia_percentual):
    capacidade_total = 1000
    consumo_decolagem = 300
    perdas = 50

    energia_disponivel = capacidade_total * (energia_percentual / 100)
    autonomia = energia_disponivel - consumo_decolagem - perdas

    return autonomia
```

## 5. Execução do sistema

Agora o sistema gera os dados, aplica as verificações e apresenta o resultado final.

```python
dados = gerar_dados()
resultado, motivos = avaliar_lancamento(dados)
autonomia = calcular_autonomia(dados["energia"])

print("=== DADOS DA NAVE ===")
for chave, valor in dados.items():
    print(f"{chave}: {valor}")

print("\n=== RESULTADO ===")
print(resultado)

for motivo in motivos:
    print("-", motivo)

print("\n=== ANÁLISE ENERGÉTICA ===")
print(f"Autonomia restante: {autonomia:.2f} kWh")
```

## 6. Observação final

O notebook pode ser executado várias vezes para gerar cenários diferentes e observar quando a nave está pronta para decolar ou quando a decolagem deve ser abortada.
